#### Problem 1(b)

Summarize the Salaries DataFrame to show the total salaries for each team for each year. Show the head of the new summarized DataFrame. 

```python
#your code here
sal = salaries.groupby(['yearID','teamID']).sum()
sal.head()
```

#### Problem 1(c)

Merge the new summarized Salaries DataFrame and Teams DataFrame together to create a new DataFrame
showing wins and total salaries for each team for each year year. Show the head of the new merged DataFrame.

**Hint**: Merge the DataFrames using `teamID` and `yearID`.

```python
#your code here
# create wins dataframe
wins = teams[['yearID','teamID','W']].groupby(['yearID','teamID']).sum()

# reset indexes of dataframe for merge
left = sal.reset_index()
right = wins.reset_index()

# merge data frames and then set the indexes once again
merged = pd.merge(left,right,on=['yearID','teamID'],how='left')
merged = merged.set_index(['yearID','teamID'])
merged.head()
```

#### Problem 1(d)

How would you graphically display the relationship between total wins and total salaries for a given year? What kind of plot would be best? Choose a plot to show this relationship and specifically annotate the Oakland baseball team on the on the plot. Show this plot across multiple years. In which years can you detect a competitive advantage from the Oakland baseball team of using data science? When did this end? 

**Hints**: Use a `for` loop to consider multiple years. Use the `teamID` (three letter representation of the team name) to save space on the plot.  

```python
#your code here
years = merged.index.levels[0]
df = merged.reset_index()

def make_plot(year):
    # select data for specific year
    data = df[df['yearID'] == year]
    data = data.reset_index(drop=True)
    x, y, names = data['salary']/1000000, data['W'], data['teamID']
    
    # create scatter plot with invisible points
    plt.scatter(x,y,color='white')

    # annotate points with teamIDs 
    for i, name in enumerate(names):
        if name == 'OAK':
            plt.annotate(name, (x[i],y[i]), color = 'green')
        else:
            plt.annotate(name, (x[i],y[i]))

    plt.suptitle(year, fontsize=14)
    plt.xlabel('Salary in Millions')
    plt.ylabel('Wins')

# plot for years 2000 and beyond
for year in years[15::]:    
    plt.show(make_plot(year))
```

The best plot to use to graphically display the relationship between total wins and total salaries is a scatter plot. From years 2000 to 2003, the Oakland Athletics seem to have had a competitive advantage by using data science. The scatter plots for these years show that the A's managed to win surprisingly more games than many of the teams that had higher team salaries. After 2004, this competitive advantage seemed to have disappeared.

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
#your code here
def mergeByYear(year):
    right = income.T[['gdp pc test',year]].set_index('gdp pc test')

    df = countries.join(right,on='Country',how='left')
    df.rename(columns={year:'Income'}, inplace=True)
    return df
```

#### Problem 2(d) 

Use exploratory data analysis tools such as histograms and boxplots to explore the distribution of the income per person by region data set from 2(c) for a given year. Describe how these change through the recent years?

**Hint**: Use a `for` loop to consider multiple years.  

```python
#your code here
def build_plots(year):
    # build data set
    yr = mergeByYear(year)
    # list to build boxplot later
    boxes = []
    
    # separate data set by region
    western = yr[(yr['Region'] == 'NORTH AMERICA') | (yr['Region'] == 'EUROPE')]
    nonwestern = yr[(yr['Region'] == 'SOUTH AMERICA') | (yr['Region'] == 'AFRICA') | (yr['Region'] == 'ASIA')]
    regions = [western,nonwestern]
    region_name = ["N Amer, Eur","S Amer, Afr, Asia"]
    
    # get list of incomes and plot histogram for the region in specified year
    for i, region in enumerate(regions):
        incomes = list(region['Income'][region['Income'].notnull()])
        histo = plt.hist(incomes)
        plt.suptitle(region_name[i] + ' in ' + str(year))
        plt.xlabel('GDP/Capita')
        plt.ylabel('Number of Countries')
        plt.show(histo)
        
        # appends data to be used later for boxplot
        boxes.append(incomes)
    
    # create box plot
    box = plt.boxplot(boxes,vert=0)
    plt.suptitle(str(year))
    plt.xlabel('GDP/Capita')
    plt.yticks([1, 2], region_name)
    plt.show(box)

# choose years for last two decades
recent_yrs = income.index[194::]
    
# build histograms and boxplots for last two decades
for yr in recent_yrs:
    build_plots(yr)
```

Over the last two decades, the distribution of income/person in the non-Western world hasn't changed much with most countries with GDP/capita less than 10,000. #### Problem 3(a)

Assume you have two list of numbers, X and Y, with distribution approximately normal. X and Y have standard deviation equal to 1, but the average of X is different from the average of Y.  If the difference in the average of X and the average of Y is larger than 0, how does the proportion of X > a compare to the proportion of Y > a? 

Write a function that analytically calculates the ratio of these two proportions: Pr(X > a)/Pr(Y > a) as function of the difference in the average of X and the average of Y. 

**Hint**: Use the `scipy.stats` module for useful functions related to a normal random variable such as the probability density function, cumulative distribution function and survival function. 

**Update**: Assume Y is normally distributed with mean equal to 0. 

Show the curve for different values of a (a = 2,3,4 and 5).

```python
"""
Function
--------
ratioNormals

Return ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a) as function of 
    the difference in the average of X 
    and the average of Y. 

Parameters
----------
diff : difference in the average of X 
    and the average of Y. 
a : cutoff value

Returns
-------
Returns ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a)
    
Example
-------
>>> ratioNormals(diff = 1, a = 2)
"""
#your code here
def ratioNormals(diff, a):
    import scipy.stats
    x_prob = 1 - scipy.stats.norm.cdf(a-diff)
    y_prob = 1 - scipy.stats.norm.cdf(a)
    
    return x_prob/y_prob
```

```python
#your code here
# define x axis and a parameter
x_list = range(-5,6)
a_list = [2,3,4,5]

# plot for different a values
for a in a_list:
    y_list = []
    for x in x_list:
        y_list.append(ratioNormals(x,a))
    
    curve = plt.plot(x_list,y_list)
    plt.suptitle('a = ' + str(a), size = 14)
    plt.xlabel('Difference in Average (avg of x - avg of y)')
    plt.ylabel('Ratio (Pr(X>a) / Pr(Y>a))')
    plt.show(curve)

```

If the difference in the average of X and the average of Y is larger than 0, how does the proportion of X > a compare to the proportion of Y > a?

**Jacob's Answer: **
If the difference in the average of X and the average of Y is larger than 0, then the proportion of X > a is larger than the proportion of Y > a. This is true because it is given that X and Y are approximately normally distributed. We can clearly see this in the plots above as well. If the difference in averages (average of x - average of y) is positive, then the ratio is greater than 1. If it is negative, then the ratio is a fraction. Another pattern that we can see is that if *a* is bigger, than the ratio increases much faster.
